## Loading Packages

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Load in Data

```python
df = pd.read_csv("https://raw.githubusercontent.com/Stephen-sanchez1/datasci_3_eda/main/datasets/diabetes_prediction_dataset.csv")
df.head(10)
```

```python
print(df.columns)
```

## 1. Univariate Analysis:
* Manually perform a univariate analysis to understand the distribution of each variable. This includes calculating measures of central tendency (mean, median, mode) and measures of spread (range, variance, standard deviation, IQR).
* Visualize the distribution of select numerical variables using histograms.


```python
# # Univariate Analysis for bllod_glucose_level
df['blood_glucose_level'] = pd.to_numeric(df['blood_glucose_level'], errors='coerce')

mean_bgl = np.mean(df['blood_glucose_level'])
median_bgl = np.median(df['blood_glucose_level'])
mode_bgl = df['blood_glucose_level'].mode().values[0]
range_bgl = df['blood_glucose_level'].max() - df['blood_glucose_level'].min()
variance_bgl = np.var(df['blood_glucose_level'])
std_deviation_bgl = np.std(df['blood_glucose_level'])
q75_bgl, q25_bgl = np.percentile(df['blood_glucose_level'], [75, 25])
iqr_bgl = q75_bgl - q25_bgl

print(f"Blood Glucose Level - Univariate Analysis:")
print(f"Mean: {mean_bgl}")
print(f"Median: {median_bgl}")
print(f"Mode: {mode_bgl}")
print(f"Range: {range_bgl}")
print(f"Variance: {variance_bgl}")
print(f"Standard Deviation: {std_deviation_bgl}")
print(f"IQR: {iqr_bgl}")
```

```python
# Univariate Analysis for HbA1c_level
df['HbA1c_level'] = pd.to_numeric(df['HbA1c_level'], errors='coerce')

mean_HbA1c = np.mean(df['HbA1c_level'])
median_HbA1c = np.median(df['HbA1c_level'])
mode_HbA1c = df['HbA1c_level'].mode().values[0]
range_HbA1c = df['HbA1c_level'].max() - df['HbA1c_level'].min()
variance_HbA1c = np.var(df['HbA1c_level'])
std_deviation_HbA1c = np.std(df['HbA1c_level'])
q75_HbA1c, q25_HbA1c = np.percentile(df['HbA1c_level'], [75, 25])
iqr_HbA1c = q75_HbA1c - q25_HbA1c

print(f"HbA1c Level - Univariate Analysis:")
print(f"Mean: {mean_HbA1c}")
print(f"Median: {median_HbA1c}")
print(f"Mode: {mode_HbA1c}")
print(f"Range: {range_HbA1c}")
print(f"Variance: {variance_HbA1c}")
print(f"Standard Deviation: {std_deviation_HbA1c}")
print(f"IQR: {iqr_HbA1c}")
```

```python
# Histogram for 'blood_glucose_level'
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='blood_glucose_level', bins=20, kde=True)
plt.title('Distribution of Blood Glucose Level')
plt.xlabel('Blood Glucose Level')
plt.ylabel('Frequency')
plt.show()
```

```python
# Histogram for 'HbA1c_level'
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='HbA1c_level', bins=20, kde=True)
plt.title('Distribution of HbA1c Level')
plt.xlabel('HbA1c Level')
plt.ylabel('Frequency')
plt.show()
```

## 2. Bivariate Analysis:
* Analyze the relationship between pairs of variables.
  * Use scatter plots to explore potential relationships between two numerical variables.
* For categorical and numerical variable pairs, use boxplots.
Compute correlation coefficients for numerical variables and document any strong correlations observed.

```python
# Scatter plot for blood_glucose_level vs HbA1c_level
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='blood_glucose_level', y='HbA1c_level')
plt.title('Relationship between Blood Glucose Level and HbA1c Level')
plt.xlabel('Blood Glucose Level')
plt.ylabel('HbA1c Level')
plt.show()
```

```python
# Compute and display correlation matrix
correlation_matrix = df.corr()
print(correlation_matrix)

# If you wish to visualize the correlations
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

## 3. Handling Outliers:
* Identify outliers in your dataset using the IQR method or visualization tools.

```python
# Step 1: Calculating IQR
Q1 = df['HbA1c_level'].quantile(0.25)
Q3 = df['HbA1c_level'].quantile(0.75)
IQR = Q3 - Q1

# Identifying Outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['HbA1c_level'] < lower_bound) | (df['HbA1c_level'] > upper_bound)]

print("Outliers:")
print(outliers)
```

```python
# Printing the number of outliers
print(f"Number of outliers in HbA1c_level: {len(outliers)}")
```

Retaining Outliers

I have decided to retain the outliers found. These outliers can represent individuals with extremely high HbA1c levels and are not errors within the dataset. They can provide valuable insights into risk factors of high HbA1c levels.
